```python
over_40_target = data1[(data1["age"] > 40) & (data1["slope"] == 1)]
over_40_target.tail()
```

```python
data1["thal"].nunique()
```

```python
fig, ax = plt.subplots()
over_40_target.plot(x="age", y="chol", c="thal", kind="scatter", ax=ax,cmap="winter");
```

```python
plt.scatter(over_40_target["age"], over_40_target["chol"], c = over_40_target["thal"])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
base = ax.scatter(x=over_40_target["age"], y=over_40_target["chol"], c=over_40_target["thal"],cmap="winter")
ax.set(title="Health Status",
       xlabel="Age",
       ylabel="Cholesterol Level");
ax.legend(*base.legend_elements(), title="Thalassemia");
ax.axhline(over_40_target["chol"].mean(),
           ls= "--",
           c= "red");
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,8),sharex=True)
brace = ax1.scatter(x=over_40_target["age"], y=over_40_target["chol"], c=over_40_target["thal"],cmap="summer")
ax1.set(title="Health Status",
       xlabel="Age",
       ylabel="Cholesterol Level");
ax1.legend(*brace.legend_elements(), title="Thalassemia");
ax1.axhline(over_40_target["chol"].mean(),
           ls= "--",
           c= "red");
brace = ax2.scatter(x=over_40_target["age"], y=over_40_target["trestbps"], c=over_40_target["thal"])
ax2.set(title="Health Status",
       xlabel="Age",
       ylabel="Cholesterol Level");
ax2.legend(*brace.legend_elements(), title="Thalassemia");
ax2.axhline(over_40_target["trestbps"].mean(),
           ls= "-",
           c= "blue");
fig.suptitle("Health Analysis", fontsize=18, fontweight="bold", horizontalalignment="center");
```
